Se crea la función principal que contiene la lógica para determinar jugadores similares a otro. Recibe como parámetro el ID del jugador a consultar y la cantidad de jugadores similares deseados

```
jugadores_similares <- function(x,y){
  
  id_jugador = x #"id" de jugador a investigar
  
  #calcula el "id_row" de un "id" de jugador dado
  nro_row = which(VectorDF$int_player_id == id_jugador) 
  
  #determina la posicion de un "id" de jugador dado
  posicion= VectorDF$str_best_position[nro_row]
  
  #factor por el que se van a multiplicar las variables dependiendo de la posicion
  factor = 10
  
  if(posicion==1) { #si es arquero
    VectorDF = mutate(VectorDF,
                      int_diving = case_when("str_best_position"==1 ~ int_diving * factor,TRUE ~ int_diving),
                      int_handling = case_when("str_best_position"==1 ~ int_handling * factor,TRUE ~ int_handling),
                      int_kicking = case_when("str_best_position"==1 ~ int_kicking * factor,TRUE ~ int_kicking),
                      int_gk_positioning = case_when("str_best_position"==1 ~ int_gk_positioning * factor,TRUE ~ int_gk_positioning),
                      int_reflexes = case_when("str_best_position"==1 ~ int_reflexes * factor,TRUE ~ int_reflexes)
    ) #pondera los atributos de los arqueros
  } else {
    if(posicion>=10 && posicion<=35) { #si es defensor
      VectorDF = mutate(VectorDF,
                        int_defensive_awareness = case_when("str_best_position">=10 && "str_best_position"<=35 ~ int_defensive_awareness * factor,TRUE ~ int_defensive_awareness),
                        int_standing_tackle = case_when("str_best_position">=10 && "str_best_position"<=35 ~ int_standing_tackle * factor,TRUE ~ int_standing_tackle),
                        int_sliding_tackle = case_when("str_best_position">=10 && "str_best_position"<=35 ~ int_sliding_tackle * factor,TRUE ~ int_sliding_tackle),
                        int_interceptions = case_when("str_best_position">=10 && "str_best_position"<=35 ~ int_interceptions * factor,TRUE ~ int_interceptions),
                        int_heading_accuracy = case_when("str_best_position">=10 && "str_best_position"<=35 ~ int_heading_accuracy * factor,TRUE ~ int_heading_accuracy)
      ) #pondera los atributos de los defensores
    } else {
      if(posicion>=40 && posicion<=70) { #si es mediocampista
        VectorDF = mutate(VectorDF,
                          int_positioning = case_when("str_best_position">=40 && "str_best_position"<=70 ~ int_positioning * factor,TRUE ~ int_positioning),
                          int_vision = case_when("str_best_position">=40 && "str_best_position"<=70 ~ int_vision * factor,TRUE ~ int_vision),
                          int_long_passing = case_when("str_best_position">=40 && "str_best_position"<=70 ~ int_long_passing * factor,TRUE ~ int_long_passing),
                          int_ball_control = case_when("str_best_position">=40 && "str_best_position"<=70 ~ int_ball_control * factor,TRUE ~ int_ball_control),
                          int_fk_accuracy = case_when("str_best_position">=40 && "str_best_position"<=70 ~ int_fk_accuracy * factor,TRUE ~ int_fk_accuracy),
                          int_short_passing = case_when("str_best_position">=40 && "str_best_position"<=70 ~ int_short_passing * factor,TRUE ~ int_short_passing),
        ) #pondera los atributos de los mediocampistas
      } else {
        VectorDF = mutate(VectorDF, #si es delantero
                          
                          int_finishing = case_when("Position">70 ~ int_finishing * factor,TRUE ~ int_finishing),
                          int_volleys = case_when("Position">70 ~ int_volleys * factor,TRUE ~ int_volleys),
                          int_dribbling = case_when("Position">70 ~ int_dribbling * factor,TRUE ~ int_dribbling),
                          int_agility = case_when("Position">70 ~ int_agility * factor,TRUE ~ int_agility),
                          int_shot_power = case_when("Position">70 ~ int_shot_power * factor,TRUE ~ int_shot_power)
        ) #pondera los atributos de los delanteros
      }
    }
  }
  
  
  
  #Calcula las componentes principales sin el "id" utilizando metodo de covarianza
  PRC <- prcomp(VectorDF[,(3:41)],scale = TRUE)
  #PRC <- prcomp(VectorDF[,(3:41)])
  
  #PRC$x
  #summary(PRC)
  
  #Calculo la variabilidad para cada componente principal
  #variabilidadCov = PRC$sdev^2/ sum(PRC$sdev^2)
  #round(variabilidadCov,2) 
  #plot(variabilidadCov, type = "o", lwd=2 , col = 'blue',main="Variabilidad por Componentes según Covarianza",xlab="Cant de Componentes",ylab="Variabilidad")
  
  
  #Agrega al dataframe los valores del PCA
  VectorDF = cbind(VectorDF,PRC$x)
  
  #selecciones el "id_row"+id"+los primeros n PCA (97% de variabilidad de los datos)
  VectorDF <- subset(VectorDF, select = c(1:2, PC1:PC31))
  
  #Agrega la columna distancia con el valor de la dist euclediana de todos los jugadores respecto a un "id_row" dado 
  VectorDF$Dist = as.matrix(dist(VectorDF[,3:33]))[nro_row, ]
  
  #Selecciona los y jugadores mas similares
  SimilaresDF = top_n(VectorDF, (-1*(1+y)), Dist) 
  
  #Busca la info de los jugadores mas similares
  SimilaresDF = merge(SimilaresDF, FifaDF, by = "int_player_id")
  
  #Devuelve los y jugadores mas similares ordenados por la distancia
  resultadoDF = select(arrange(SimilaresDF,Dist), int_player_id, Dist, str_player_name, dt_date_of_birth,str_nationality,str_club_name,str_league_name,int_value,int_wage)
  resultadoDF=rename(resultadoDF, ID = int_player_id,NAME=str_player_name,DATE_OF_BIRTH=dt_date_of_birth,NATIONALITY=str_nationality,CLUB=str_club_name,LIGA=str_league_name,VALUE=int_value,WAGE=int_wage)
  return(resultadoDF)
}
```
